````r
# PASO 5: Manejo de Valores Faltantes (Imputación por la media)
# Calcular la media de la columna 'Resistencia' excluyendo los NA
media_resistencia <- mean(datos_r$Resistencia, na.rm = TRUE)

# Imputar los valores NA en 'Resistencia' con la media calculada
datos_r$Resistencia[is.na(datos_r$Resistencia)] <- media_resistencia

print("Valores NA por columna después de la imputación:")
print(colSums(is.na(datos_r)))

print("Resumen Estadístico después de la imputación:")
summary(datos_r)
````
